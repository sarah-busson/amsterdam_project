```python
#And then drop the columns where all the values are null.
```

```python
null_data = nl_data_re.isnull().sum()
null_data[null_data == 55]
 
nl_data_re = nl_data_re.drop(list(null_data[null_data == 55].index), axis=1)
nl_data_re.head()
```

```python
#Also, we are interested in the data from 1990 to the most recent
```

```python
nl_data_re = nl_data_re[nl_data_re['Year'] >= 1990]
nl_data_re.head()
```

```python
#Alright, so now let's harmonize the data and keep only the one type in one type of unit.
nl_data_re.columns
#The most common unit is EJ (exajoule) so let's keep those columns 
#and see what we can do with the other ones.
```
